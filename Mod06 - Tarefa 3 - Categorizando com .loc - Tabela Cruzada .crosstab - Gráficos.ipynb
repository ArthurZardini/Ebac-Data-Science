#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df.loc[df['APGAR5'] <= 3, 'APGAR5_cat'] = 'asfixia grave'
df.loc[(df['APGAR5'] >= 4) & (df['APGAR5'] <= 5), 'APGAR5_cat'] = 'asfixia moderada'
df.loc[(df['APGAR5'] >= 6) & (df['APGAR5'] <= 7), 'APGAR5_cat'] = 'asfixia leve'
df.loc[df['APGAR5'] >= 8, 'APGAR5_cat'] = 'normal'

df
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.boxplot(data = df, y = "GESTACAO", x= "APGAR5", hue="SEXO")

```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
df_cross = pd.crosstab(df['GESTACAO'], df['APGAR5'])
df_cross
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
cross_perc = df_cross.div(df_cross.sum(axis =1), axis = 0)
cross_perc
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
cross_perc.sum(axis = 1)
```

```python
cross_perc.plot.bar()
```

De acordo com o gráfico acima, notamos que os níveis de asfixia sobem nas crianças com menos de 37 semanas.
E conforme o número de semanas vai diminuindo, o número de crianças com asfixia grave vai aumentando.
Ou seja, quanto menos tempo a criança passa na barriga da mãe, maior é a chance dela ter algum tipo de asfixia ao nascer.
