A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

Média e evolução:

```python
#a
df['média'] = df.mean(axis=1)
df['evolução'] = df.p2-df.p1
df.style.format(precision=1, decimal=',')
```

```python
#b
correlacao_p1_p2 = df['p1'].corr(df['p2'])
print("Correlação entre p1 e p2:", correlacao_p1_p2)
```

```python
#c
correlacao_media_evolucao = df['média'].corr(df['evolução'])
print("Correlação entre média e evolução:", correlacao_media_evolucao)
```

d)
- Correlação entre p1 e p2: <br>
É uma correlação forte, de quase 90%.<br>
Por exemplo, quando os valores de p1 aumentam, os valores de p2 tendem a aumentar também.<br>
<br>
- Correlação entre Média e Evolução:<br>
É uma correlação bem baixa, bem fraca, de apenas 13.34%.<br>
Ou seja, Média e Evolução são independentes, não tem muita correlação. Caso a Média suba, não quer dizer que a Evolução vai subir.


## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

```python
df['média'] = df.mean(axis=1)
df['soma'] = df.p1+df.p2
df.style.format(precision=1, decimal=',')
```
