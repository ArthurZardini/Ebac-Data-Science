## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
txt = 'Roda mundo, roda gigante, roda moinho, roda pião.'

qtd_repeticao_roda = txt.count("roda")
print(qtd_repeticao_roda)

split_gigante = txt.split("gigante")
print(split_gigante)

termina_ponto_final = txt.endswith(".")
print(termina_ponto_final)

```
